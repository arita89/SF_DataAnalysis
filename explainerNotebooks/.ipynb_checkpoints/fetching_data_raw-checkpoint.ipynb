# edit sf_weather_data

```python
sf_weather_data = pd.read_csv("../data/sf-20082021/weather_data_1hr.csv")
#sf_weather_data['time'] = sf_weather_data['time'].apply(str)
```

```python
sf_weather_data["hourOfDay"]= (sf_weather_data["time"]/100).apply(int)
sf_weather_data = resetColumns(sf_weather_data,frompos = -1,insertAt=1)
```

```python
sf_weather_data.columns
```

```python
w = sf_weather_data[['date', 'hourOfDay', 'isdaytime', 
                     'tempC','humidity','pressureMB','precipMM',
                     'cloudcover', 'HeatIndexC','uvIndex',
                     'windspeedKmph','winddirdegree',
                     'weatherCode','weatherDesc'
                    ]]
w
```

```python
w = addDate(w,"date")
w
```

```python
def formatWData(df,insertAt=1, hourshift = 0):
    df['NumDayOfWeek'] = df['Date'].dt.dayofweek
    #df['Date'] = df['Date'].dt.date
    df = resetColumns(df,frompos = -1,insertAt=1)
    return df 
```

```python
wfor = formatWData(w)
```

```python
wfor
```

```python
start_date = "01/07/2008" # mm/dd/yyyy
end_date = "01/01/2021"
mask = (wfor['Date'] >= start_date) & (wfor['Date'] < end_date)
wformasked = wfor.loc[mask].sort_values(by=['date',"hourOfDay"], inplace=False, ascending=True)
```

```python
wformasked
```

```python
wformasked = wformasked[['Date', 'NumDayOfWeek', 'hourOfDay', 'isdaytime', 'tempC',
       'humidity', 'pressureMB', 'precipMM', 'cloudcover', 'HeatIndexC',
       'uvIndex', 'windspeedKmph', 'winddirdegree', 'weatherDesc']]
```

```python
wformasked.reset_index(inplace=True,drop=True)
```
